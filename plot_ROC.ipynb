```python
import pickle

# Open the .pkl file in read-binary mode
t = 4
with open(f'/home/pyq6817/ROC_curve_data_Train_Test_{t}.pkl', 'rb') as file:
    data = pickle.load(file)

# Now, `data` contains the content of the .pkl file
print(len(data))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
for i in range(5):
    fold_data = data[i]
    fprs = fold_data[0]
    tprs = fold_data[1]

    plt.plot(fprs, tprs, linewidth=2,label=f'Fold {i+1} (AUC={fold_data[2]:.2f})')
    
# Add diagonal line (Random classifier)
plt.plot([0, 1], [0, 1], color='black', linestyle='--', linewidth=1)

# Add labels and title
plt.title(f'T{t}')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend(loc='lower right')

# Display the plot
plt.grid(True)
plt.show()
```
